```python
import numpy as np
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.preprocessing import MinMaxScaler, StandardScaler
boston=load_boston()
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df['Price']=boston.target
```

```python
boston_df._get_numeric_data()
```

```python
s_scaler = StandardScaler()
s_scaler.fit_transform(boston_df._get_numeric_data())[0:2] 
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit_transform(boston_df._get_numeric_data())[0:2]
```

```python
sns.barplot(data=boston.target)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=boston_df, x="AGE", y="Price")
plt.show()
```

```python
sns.catplot('Price', data=boston_df, kind="box")
plt.show()
```
